```python
import numpy as np
import matplotlib.pyplot as plt
import os
import torch
import torchvision as tv
import random
%matplotlib qt
```

```python
window_size = 20
numsplits = 20
reload = True
modes = ['train','test']

class_folders = [
            ['ball_data'],
            ['bott_data'],
            ['fist_data'],
            ['ind_data'],
            ['mid_data'],
            ['rin_data'],
            ['pin_data'],
            ['thr_data']
]
new_folder = 'combined_data'
```

```python
if reload:
    for mode in modes:
        for i in range(numsplits):
            np.save('{0}/data_{1}_{2}.npy'.format(new_folder,i,mode),np.zeros((0,636*256*8+1)))
```

```python
for mode in modes:
    files = [(i,os.path.join(d,f)) for i,l in enumerate(class_folders) for d in l for f in os.listdir(d) if f.endswith('-{0}-images.npy'.format(mode))]
    random.shuffle(files)

    for i,file in enumerate(files):
        data = np.load(file[1])
        datalen = data.shape[0]-window_size+1
        splits = np.array_split(np.arange(datalen),numsplits,axis=0)
        random.shuffle(splits)
        print(file[1])
        for j,split in enumerate(splits):
            print('file {0} slice {1} loading'.format(i,j),end=' ')
            iname = '{0}/data_{1}_{2}.npy'.format(new_folder,j,mode)
            previous = np.load(iname)

            prevlen = previous.shape[0]
            newlen = split.shape[0]
            print('{0} items adding {1} items'.format(prevlen,newlen),end=' ')

            newdata = np.zeros((prevlen+newlen,636*256*8+1))
            newdata[0:prevlen,:] = previous
            for idx in split:
                newdata[prevlen+idx:,1:] = np.moveaxis(data[idx:idx+window_size,...],0,-1).flatten()
            newdata[prevlen:,0] = file[0]*np.ones((newlen,))
            print('saving')
            np.save(iname,newdata)
            del previous
            del newdata
        print(file[1],'finished')
        del data
```
